```python
strike_dip1 = [[290,46],[281,47],[283,39],[285,45],[287,37],[299,51]]
strike_dip2 = [[325,38],[304,37],[308,32],[314,32],[318,34]]
strike_dip3 = [[347,46],[355,33],[5,38]]
strike_dip4 = [[330,44],[344,40],[335,37]]
strike_dip5 = [[3,47],[1,37],[352,46],[4,42]]
strike_dip6 = [[311,46],[325,47],[314,36]]
```

```python
bedding_data1 = pd.DataFrame(strike_dip1,columns=['strike','dip'])
bedding_data2 = pd.DataFrame(strike_dip2,columns=['strike','dip'])
bedding_data3 = pd.DataFrame(strike_dip3,columns=['strike','dip'])
bedding_data4 = pd.DataFrame(strike_dip4,columns=['strike','dip'])
bedding_data5 = pd.DataFrame(strike_dip5,columns=['strike','dip'])
bedding_data6 = pd.DataFrame(strike_dip6,columns=['strike','dip'])
```

```python
bedding_data1['pole_trend'] = bedding_data1['strike']-90
bedding_data1['pole_plunge'] = 90 - bedding_data1['dip']
bedding_data2['pole_trend'] = bedding_data2['strike']-90
bedding_data2['pole_plunge'] = 90 - bedding_data2['dip']
bedding_data3['pole_trend'] = bedding_data3['strike']-90
bedding_data3['pole_plunge'] = 90 - bedding_data3['dip']
bedding_data4['pole_trend'] = bedding_data4['strike']-90
bedding_data4['pole_plunge'] = 90 - bedding_data4['dip']
bedding_data5['pole_trend'] = bedding_data5['strike']-90
bedding_data5['pole_plunge'] = 90 - bedding_data5['dip']
bedding_data6['pole_trend'] = bedding_data6['strike']-90
bedding_data6['pole_plunge'] = 90 - bedding_data6['dip']
```

The best way to take the average of planes is to calculate the poles to each plane and take a mean of that "Fisherian" distribution. The plane perpendicular to the mean is the best average of the measured bedding planes.

```python
bedding_poles1 = ipmag.make_di_block(bedding_data1['pole_trend'],bedding_data1['pole_plunge'])
bedding_poles_mean1 = pmag.fisher_mean(bedding_poles1)
bedding_poles2 = ipmag.make_di_block(bedding_data2['pole_trend'],bedding_data2['pole_plunge'])
bedding_poles_mean2 = pmag.fisher_mean(bedding_poles2)
bedding_poles3 = ipmag.make_di_block(bedding_data3['pole_trend'],bedding_data3['pole_plunge'])
bedding_poles_mean3 = pmag.fisher_mean(bedding_poles3)
bedding_poles4 = ipmag.make_di_block(bedding_data4['pole_trend'],bedding_data4['pole_plunge'])
bedding_poles_mean4 = pmag.fisher_mean(bedding_poles4)
bedding_poles5 = ipmag.make_di_block(bedding_data5['pole_trend'],bedding_data5['pole_plunge'])
bedding_poles_mean5 = pmag.fisher_mean(bedding_poles5)
bedding_poles6 = ipmag.make_di_block(bedding_data6['pole_trend'],bedding_data6['pole_plunge'])
bedding_poles_mean6 = pmag.fisher_mean(bedding_poles6)
```

## Tilt correction for set_1 (bottom of U1424)

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='stereonet')

mean_strike1 = bedding_poles_mean1['dec'] + 90.0
mean_dip1 = 90.0 - bedding_poles_mean1['inc']

ax.plane(bedding_data1['strike'],bedding_data1['dip'], 'g-', linewidth=1.5)
ax.pole(bedding_data1['strike'],bedding_data1['dip'], 'bo', markersize=8)
ax.plane(mean_strike1,mean_dip1, 'r-', linewidth=3)
ax.pole(mean_strike1,mean_dip1, 'r^', markersize=15)
ax.grid()

plt.show()
print mean_strike1,mean_dip1
```
